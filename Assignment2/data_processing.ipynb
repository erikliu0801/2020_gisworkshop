```python
def chinese_to_int(text):
    if type(text) != str: return text
    digits = [0 ,'十','百'] #個位數不用存
    numbers = [0, '一', '二', '三', '四', '五', '六', '七', '八', '九']
    num = 0
    text = text[::-1] #倒著計算
    
    digit = 0
    for t in text:
        if t in digits:
            digit = digits.index(t)
            continue
        if t in numbers:
            num += numbers.index(t) * (10 ** digit)
            digit = 0
            continue
    if digit != 0:
        num += 10 ** digit
    return num
```

```python
for i, f in enumerate(floor):
    if f:
        floor[i] = chinese_to_int(f)
    print(floor[i])
```
